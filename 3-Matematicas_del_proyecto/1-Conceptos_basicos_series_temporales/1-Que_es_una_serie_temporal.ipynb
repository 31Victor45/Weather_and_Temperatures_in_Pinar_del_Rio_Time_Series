## ¬øQu√© es una Serie de Tiempo en Python? ‚è≥

Una serie de tiempo (time series) es una secuencia de puntos de datos indexados, listados o graficados en orden cronol√≥gico. Es decir, es un conjunto de observaciones sobre un fen√≥meno tomadas en momentos sucesivos y espaciados de manera uniforme o irregular.

Ejemplos Comunes:

1. Precios diarios de acciones en la bolsa.

2. Temperaturas mensuales promedio en una ciudad.

3. Volumen de ventas por hora de un producto.

3. Registros de tr√°fico web por minuto.

## Estructura en Python (Pandas)
En Python, la librer√≠a fundamental para trabajar con series de tiempo es Pandas.

Una serie de tiempo se representa t√≠picamente como un objeto Series o DataFrame de Pandas, donde la caracter√≠stica clave es el uso de un DatetimeIndex (√≠ndice de fecha y hora).

Creaci√≥n de una Serie de Tiempo b√°sica en Pandas
El siguiente c√≥digo muestra c√≥mo se crea una serie simple con un √≠ndice de tiempo:

```python
import pandas as pd
import numpy as np

# 1. Crear un √≠ndice de tiempo (fechas diarias por 5 d√≠as)
fechas = pd.date_range(start='2025-01-01', periods=5, freq='D')

# 2. Crear datos aleatorios
valores = np.random.randint(10, 100, size=5)

# 3. Crear la Serie de Tiempo
serie_tiempo = pd.Series(valores, index=fechas)

# Mostrar la serie
print("Tipo de Objeto:", type(serie_tiempo))
print("Tipo de √çndice:", type(serie_tiempo.index))
print("\nSerie de Tiempo:")
print(serie_tiempo)
```
